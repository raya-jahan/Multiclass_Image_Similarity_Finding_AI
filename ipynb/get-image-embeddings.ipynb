```python
import os
import base64
import math
from io import BytesIO
from PIL import Image
from pandas import DataFrame
from dotenv import load_dotenv
from transformers import AutoImageProcessor, ResNetForImageClassification
from qdrant_client import QdrantClient, models
from qdrant_client.models import VectorParams, Distance

# Load environment variables
load_dotenv()
qclient = QdrantClient(
    url=os.getenv('QDRANT_DB_URL'),
    api_key=os.getenv('QDRANT_API_KEY')
)

# Load cat images
cat_dir = "./afhq/train/cat"
sample_cat_image_paths = [
    os.path.join(cat_dir, fname) for fname in os.listdir(cat_dir)[:60]
]

# Resize & encode
def resize_image(image_path):
    img = Image.open(image_path)
    aspect = img.width / img.height
    return img.resize([256, math.floor(256 * aspect)])

def convert_image_to_base64(img):
    buffer = BytesIO()
    img.save(buffer, format="JPEG")
    return base64.b64encode(buffer.getvalue()).decode("utf-8")

resized_images = list(map(resize_image, sample_cat_image_paths))
base64_strings = list(map(convert_image_to_base64, resized_images))

# Create payloads
payloads = DataFrame({
    "image_url": "./afhq/train/cat", 
    "base64": base64_strings,
    "type": "cat"
})

# Load model & compute embeddings
processor = AutoImageProcessor.from_pretrained("microsoft/resnet-50", use_fast=True)
model = ResNetForImageClassification.from_pretrained("microsoft/resnet-50")

inputs = processor([Image.open(p) for p in sample_cat_image_paths], return_tensors="pt")
embeddings = model(**inputs).logits

# Create collection
embedding_length = embeddings.shape[-1]
collection_name = "animal_images"
qclient.recreate_collection(
    collection_name=collection_name,
    vectors_config=VectorParams(size=embedding_length, distance=Distance.COSINE)
)

# Upload records
records = [
    models.Record(
        id=idx,
        payload=payloads.iloc[idx].to_dict(),
        vector=embeddings[idx].detach().numpy().tolist()
    )
    for idx in range(len(payloads))
]

qclient.upload_records(collection_name=collection_name, records=records)
```

```python
def check_uploaded_class(label):
    results, _ = qclient.scroll(collection_name="animal_images", with_vectors=False, limit=1000)
    count = sum(1 for r in results if r.payload.get("type") == label)
    print(f"{label.upper()} Records Found: {count}")

check_uploaded_class("cat")
```

```python
import os
import base64
import math
from io import BytesIO
from PIL import Image
from pandas import DataFrame
from dotenv import load_dotenv
from transformers import AutoImageProcessor, ResNetForImageClassification
from qdrant_client import QdrantClient, models
from qdrant_client.models import VectorParams, Distance

# --- Load .env and Qdrant client
load_dotenv()
qclient = QdrantClient(
    url=os.getenv('QDRANT_DB_URL'),
    api_key=os.getenv('QDRANT_API_KEY')
)

# --- Wild image directory
wild_dir = "./afhq/train/wild"
sample_wild_image_paths = [
    os.path.join(wild_dir, fname) for fname in os.listdir(wild_dir)[:60]
]

# --- Resize and encode functions
def resize_image(image_path):
    img = Image.open(image_path)
    aspect = img.width / img.height
    return img.resize([256, math.floor(256 * aspect)])

def convert_image_to_base64(img):
    buffer = BytesIO()
    img.save(buffer, format="JPEG")
    return base64.b64encode(buffer.getvalue()).decode("utf-8")

# --- Process images
resized_wild = list(map(resize_image, sample_wild_image_paths))
base64_wild = list(map(convert_image_to_base64, resized_wild))

# --- Create payloads
wild_payloads = DataFrame({
    "image_url": sample_wild_image_paths,
    "base64": base64_wild,
    "type": "wild"
})

# --- Load model & embeddings
processor = AutoImageProcessor.from_pretrained("microsoft/resnet-50", use_fast=True)
model = ResNetForImageClassification.from_pretrained("microsoft/resnet-50")

wild_inputs = processor([Image.open(p) for p in sample_wild_image_paths], return_tensors="pt")
wild_embeddings = model(**wild_inputs).logits

# --- Upload wild image records (offset 20000)
wild_records = [
    models.Record(
        id=20000 + idx,
        payload=wild_payloads.iloc[idx].to_dict(),
        vector=wild_embeddings[idx].detach().numpy().tolist()
    )
    for idx in range(len(wild_payloads))
]

qclient.upload_records(collection_name="animal_images", records=wild_records)
```

```python
def check_uploaded_class(label):
    results, _ = qclient.scroll(collection_name="animal_images", with_vectors=False, limit=1000)
    count = sum(1 for r in results if r.payload.get("type") == label)
    print(f"{label.upper()} Records Found: {count}")

check_uploaded_class("wild")
```
